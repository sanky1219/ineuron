17. How do you distinguish between copy.copy() and copy.deepcopy()?


copy.copy() just takes the memory location of the variable been assigned but copy.deepcopy() would create a separate memory location for new variable and has its own copy. So in deepcopy even if the paarent variable is updated child variable doesn't get updated. Below example illustrates the same behaviour where id of c would be different from d but values id would be same in copy.copy() while both d id and values of d id as well is different in deepcopy.


import copy

a = [1, 2, 3]



b = [4, 5, 6]


c = [a, b]



d = copy.copy(c)



print id(c) == id(d)          --> False - d is now a new object


print id(c[0]) == id(d[0])    --> True - d[0] is the same object as c[0]


d = copy.deepcopy(c)



print id(c) == id(d)          --> False - d is now a new object


print id(c[0]) == id(d[0])    --> False - d[0] is now a new object
