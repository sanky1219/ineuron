6. What is the difference between a Run object and a Paragraph object?


The Document object contains a list of Paragraph objects for the paragraphs in the word document. (A new paragraph begins whenever the user presses ENTER or RETURN while typing in a Word document.) Each of these Paragraph objects contains a list of one or more Run objects.


A Run object is a contiguous run of text with the same style. A new Run object is needed whenever the text style changes.

8. What type of object has bold, underline, italic, strike, and outline variables?


Every run object has these all attributes with either True or False values. We can set the values to True if we want to apply or if we apply False we can remove that style, if we apply None then the default setting style is considered. 



9. What is the difference between False, True, and None for the bold variable?


Runs can be further styled using text attributes. Each attribute can be set to one of three values: True (the attribute is always enabled, no matter what other styles are applied to the run), False (the attribute is always disabled), or None (defaults to whatever the run’s style is set to).

12. What integers represent the levels of headings available in Word documents?


The arguments to add_heading() are a string of the heading text and an integer from 0 to 4. The integer 0 makes the heading the Title style, which is used for the top of the document. Integers 1 to 4 are for various heading levels, with 1 being the main heading and 4 the lowest subheading
