11. How can you tell the difference between break and continue?


Break statement exits loop and continues with the remaining statements in the program, whereas continue statement passes the loop for that corresponding element without executing the following statements and continues with next elements in the loop.


In below program for every element the loop would not print the Hello, <name> statement even the condition meets because continue is mentioned due to which python interpreter will not executing following statements and takes the control back to next element in the loop. Also once name value is Cero in the loop the program exits the for loop and executes Done! print statement, because as soon as python interpreter sees break statement, it immediately exits the loop.


names = ["Sanky", "Raju", "Machie", "Jagan", "Cero","John"]

for name in names:
    if len(name) != 4:
        continue

    print('Hello, {}'.format(name))

    if name == "Cero":
        break

print("Done!")
