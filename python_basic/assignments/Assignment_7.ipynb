5. In the regex which created from the r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group zero cover? Group 2? Group 1?


Below code illustrates the group behaviour. group(0) returns full string match, group(1) returns first sub-group match (\d\d\d) and group(2) returns second sub-group match (\d\d\d-\d\d\d\d)


import re


p = re.compile(r'(\d\d\d)-(\d\d\d-\d\d\d\d)')


m = p.match('345-123-1243')


m.group(0)    #'345-123-1243'



m.group(1)    #'345'


m.group(2)    #'123-1243'
