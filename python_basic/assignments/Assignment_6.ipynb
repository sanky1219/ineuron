```python
1. What are escape characters, and how do you use them?


Escape characters are characters that are generally used to perform certain tasks, represents certain whitespace characters and their usage in code directs the 
compiler to take a suitable action mapped to that character. Below are some escape chaaratcers which will use backslash '\'
to prefix.


'\n'  -->  Leaves a line
'\t'  -->  Leaves a space 




2. What do the escape characters n and t stand for?


'\n'  -->  Leaves a line
'\t'  -->  Leaves a space 



3. What is the way to include backslash characters in a string?


We need to prefix it with one more backslash which converts it to normal string.


a="san\tsan"


print(a) #san	san


a="san\\tsan"


print(a) #san\tsan



4. The string "Howl's Moving Castle" is a correct value. Why isn't the single quote character in the word Howl's 
not escaped a problem?


Single quote used inside the double quote would treat it as string literal rathers than as quote character. So this is the reason
it treats the single quote chaaracter as part of the string.


5. How do you write a string of newlines if you don't want to use the n character?


We cane write using 3 quotes as below


a="""Howl's Moving Castle
is a place some where 
in the world"""


print(a) #Howl's Moving Castle
#is a place some where 
#in the world 

6. What are the values of the given expressions?
'Hello, world!'[1]
'Hello, world!'[0:5]
'Hello, world!'[:5]
'Hello, world!'[3:]


e


Hello


Hello


lo, world!



7. What are the values of the following expressions?
'Hello'.upper()
'Hello'.upper().isupper()
'Hello'.upper().lower()


HELLO


True


hello


8. What are the values of the following expressions?
'Remember, remember, the fifth of July.'.split()
'-'.join('There can only one.'.split())


['Remember,', 'remember,', 'the', 'fifth', 'of', 'July.']


'There-can-only-one.'


9. What are the methods for right-justifying, left-justifying, and centering a string?


rjust, ljust and center are methods used. Below is the example that illustrates


width = 20
print('abcdefghij'.ljust(width,'-'))

output:
abcdefghij----------  


width = 20
print('abcdefghij'.rjust(width,'-'))

output:
----------abcdefghij 


width = 20
print('abcdefghij'.center(width,'-'))

output:
-----abcdefghij----- 

10. What is the best way to remove whitespace characters from the start or end?



strip method can be used. Below example illustrates


a='   abcd   '


output:
'   abcd   '



a.strip()
output: 'abcd'
```
