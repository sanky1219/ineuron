1. To what does a relative path refer?



Relative path refers to take current working directory as a reference and add the path or file mentioned to the current working directory path. We can use dot (.) and dot-dot (..) folders. These are not real folders but special names that can be used in a path. A single period (“dot”) for a folder name is shorthand for “this directory.” Two periods (“dot-dot”) means “the parent folder.”


Example:


Import os


A = Os.getcwd()


Os.path.relpath(‘.\) #Returns the current directory path.

3. What do the functions os.getcwd() and os.chdir() do?



os.getcwd() will return the current working directory(by default the program file path which is getting executed) whereas 


os.chdir() can be used to change the current working directory.

6. What are the three “mode” arguments that can be passed to the open() function?


"r" - Read - Default value. Opens a file for reading, error if the file does not exist


"a" - Append - Opens a file for appending, creates the file if it does not exist


"w" - Write - Opens a file for writing, creates the file if it does not exist


7. What happens if an existing file is opened in write mode?


Write mode overwrites the existing content and replaces it with the new content in the file, if file doesn’t exist it tries to 


create a new file and write content into it.


8. How do you tell the difference between read() and readlines()?


Read() method returns the entire content of the file as a string whereas readlines() returns a list with all the lines of the 


file as individual elements (strings).
