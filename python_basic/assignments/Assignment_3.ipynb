2. When does the code in a function run: when it's specified or when it's called? 


The function is basically registered when we run its specification and the code inside the function would run only when it is called. See below example with output that illustrates this behaviour


def fn(ka='faa'):
    return ka
    
Output: no output it is just registered in memory


sa = fn()



print(sa)

4. What is the difference between a function and a function call? 


The function is basically a specification on set of instructions on how to execute the code and return some value whereas the function call would actually execute the code inside the function and aassign the return value to the variable or any other instance whereever it is called as per the specifications.

5. How many global scopes are there in a Python program? How many local scopes? 


Global scope would be only one in python and it would be throughout the program during execution.Local scope would be specific to each function or method and could be multiple based on number of functions or methods created in the program.




6. What happens to variables in a local scope when the function call returns? 


Variable local scope would be ended once the function call returns the value and then if in case the variable is declared globally then the global scope would still be active.



7. What is the concept of a return value? Is it possible to have a return value in an expression? 


Concept of return value is to capture the output of the code in the function and assign it to the variable or other instance wherever it is called in runtime. Yes return value could be used in an expression. Below example illustrates this behaviour.


def fn(ka='faa'):
    return ka
    
Output: no output it is just registered in memory


sa = 'a' + fn()



print(sa)


output: afaa
