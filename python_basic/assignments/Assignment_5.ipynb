3. What is the most significant distinction between a dictionary and a list?


List would have index or key as only integers that in specific order starting from 0, whereas for dictionary we can have string, numeric etc. as keys.
4. What happens if you try to access spam['foo'] if spam is {'bar': 100}?


We would get a KeyError as there is no key with name foo in the spam dictionary.

5. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.keys()?


Both gives the same result either True or False based on if cat is present as a key or not. As by default python interppreter will search for the occurence of the string in the keys of that dictionary not in values



6. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?


First expression would search in dictionary keys list whereas the second expression search in dictionary values list for the string cat and return True if found else returns False.
