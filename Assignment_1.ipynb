Example: 


a = ['san', 1, '2', 3.0]


a[0] = 'san'


a[1] = 1

10. What is the error caused by this expression? What would you do about it?


'I have eaten ' + 99 + ' burritos.'


Answer: 


In above expression python interpreter throws TypeError as we are trying to concatenate string datatype values('I have eaten ', ' burritos.') to integer datatype  value(99). To avoid the error and concatenate above expression, we need to do as below by using str() function on value 99 and then concatenate with the string datatype values.



'I have eaten ' + str(99) + ' burritos.'
